```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 2: Select the survey question responses for PCA (assuming they are in columns 3 to 14)
question_responses = data.iloc[:, 2:14]  # Assuming Q1 to Q12 are in these columns

# Step 3: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(question_responses)

# Step 4: Perform PCA to extract principal components
pca = PCA()
pca_components = pca.fit_transform(data_scaled)

# Step 5: Determine the optimal number of clusters using the Silhouette Score method
silhouette_scores = []
range_n_clusters = range(2, 8)  # Try K values from 2 to 7

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(pca_components[:, :4])  # Use the first 4 principal components
    silhouette_avg = silhouette_score(pca_components[:, :4], cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Cluster Counts')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```
